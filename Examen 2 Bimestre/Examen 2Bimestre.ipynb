
## 12) Crear un Dataframe con 10 registros y 6 columnas llenas de strings. Luego, unir la columna 1 y 2 en una sola, la 3 y 4, y la 5 y 6 concatenando su texto.



````python
import string
import random
def string_aleatorio():
    ar=np.arange(6)
    st=""
    for i in ar:
        
        letra=random.choice(string.ascii_letters)
        #print(letra)
        st=st+letra
    return st
    
    
````

````python
arreglo_letras=[string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio(),string_aleatorio()]
df = pd.DataFrame({'A': np.random.choice([string_aleatorio(),string_aleatorio()],10),
                       'B':np.random.choice([string_aleatorio(),string_aleatorio()],10),
                      'C':np.random.choice([string_aleatorio(),string_aleatorio()],10),
                      'D':np.random.choice([string_aleatorio(),string_aleatorio()],10),
                      'E':np.random.choice([string_aleatorio(),string_aleatorio()],10),
                      'F':np.random.choice([string_aleatorio(),string_aleatorio()],10),})
dfr=df
dfr['A']=  df['A']+df['B']
dfr['B']=  df['C']+df['D']
dfr['C']=  df['E']+df['F']
dfr
````
